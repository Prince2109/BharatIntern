```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have a DataFrame named 'df' with columns: Open, High, Low, Close, Adj Close, and Volume

# Create a figure and axis for the plot

# Plot each column using the DataFrame and specify labels
plt.subplot(611)
plt.title("Date vs Open")
plt.plot(df['Date'], df['Open'])
plt.xlabel("Date")
plt.ylabel("Open")
plt.subplot(612)
plt.title("Date vs High")
plt.plot(df['Date'], df['High'])
plt.xlabel("Date")
plt.ylabel("High")
plt.subplot(613)
plt.title("Date vs Low")
plt.plot(df['Date'], df['Low'])
plt.xlabel("Date")
plt.ylabel("Low")
plt.subplot(614)
plt.title("Date vs Close")
plt.plot(df['Date'], df['Close'])
plt.xlabel("Date")
plt.ylabel("Close")
plt.subplot(615)
plt.title("Date vs Adj Close")
plt.plot(df['Date'], df['Adj Close'])
plt.xlabel("Date")
plt.ylabel("Adj Close")
plt.subplot(616)
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Date vs Volume")
plt.plot(df['Date'], df['Volume'])
plt.subplots_adjust(top=4.8)
# Show the plot
plt.show()
```
